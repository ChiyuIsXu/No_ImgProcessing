## Mean Filtering

```python
# 导入图像
img_bgr = cv.imread(filename=config.TEST_IMAGE, flags=cv.IMREAD_COLOR)
# img_gray = cv.imread(filename=config.TEST_img_gray, flags=cv.IMREAD_GRAYSCALE)
img_gray = cv.cvtColor(src=img_bgr, code=cv.COLOR_BGR2GRAY)

# output img_gray info
print("Shape of the color img_gray: ", img_bgr.shape)
print("Shape of the grayscale img_gray: ", img_gray.shape)
print("Data type of the img_gray: ", img_bgr.dtype)


# 显示图像
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)), plt.title('Original img_gray')
plt.subplot(1, 2, 2), plt.imshow(cv.cvtColor(img_gray, cv.COLOR_BGR2RGB)), plt.title('Gray img_gray')
plt.show()
```

### Apply Mean Filtering

```python

kernel_size = 10
mean_filtered = cv.blur(img_bgr, (kernel_size, kernel_size))

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)), plt.title('Original img_gray')
plt.subplot(1, 2, 2), plt.imshow(cv.cvtColor(mean_filtered, cv.COLOR_BGR2RGB)), plt.title('Mean Filtered img_gray')
plt.show()
```

# Median Filtering

中值滤波是一种非线性滤波技术，它通过将像素的邻域内像素值的中值替换该像素的值。
这种方法对去除椒盐噪声（随机分布的黑白噪声）特别有效，同时在保持边缘信息方面比均值滤波更好。

优点：
对椒盐噪声特别有效。
比均值滤波更好地保留图像边缘。

缺点：
计算复杂度高于均值滤波。
对于高频噪声（如高斯噪声）效果不如均值滤波。

```python
kernel_size = 11
mean_filtered = cv.medianBlur(img_bgr, kernel_size)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)), plt.title('Original img_gray')
plt.subplot(1, 2, 2), plt.imshow(cv.cvtColor(mean_filtered, cv.COLOR_BGR2RGB)), plt.title('Median Filtered img_gray')
plt.show()
```

## Gaussian Filtering

高斯滤波是一种基于高斯函数的平滑技术，它使用高斯分布的加权系数来对图像进行滤波。
相比均值滤波，高斯滤波在平滑图像时可以更好地保留边缘，因为它对邻域中心的像素赋予更高的权重。

缺点: 对于非高斯噪声效果不如其他滤波器
