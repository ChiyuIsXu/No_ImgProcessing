## interpolation 插值

插值是一种在已知数据点之间估算未知值的方法，在绘制图像时常常用来平滑曲线、填充缺失数据或者调整图像分辨率。

**线性插值：**通过在两个已知数据点之间用一条直线进行估算，简单且计算量小，适用于数据点较少的情况。

**多项式插值：**通过一个多项式函数来拟合所有数据点，常用的方法有拉格朗日插值和牛顿插值。适合数据点较多的情况，但容易出现“插值震荡”现象。

**样条插值：**使用分段的多项式函数来进行插值，其中最常见的是三次样条插值。三次样条插值在数据点之间能保持平滑的过渡，避免了多项式插值中的震荡问题。

**双线性插值：**用于二维网格数据，通过对每个方向上进行线性插值来估算值，常用于图像缩放和旋转。

**双三次插值：**一种在二维数据网格中进行插值的方法，通过对周围16个点进行加权平均来估算中间点的值，常用于图像放大时，能提供较好的平滑效果。

**高斯过程插值：**通过高斯过程来对数据进行建模，具有较好的理论基础和灵活性，但计算复杂度较高。

**克里金插值（Kriging）：**一种地质统计学方法，通过考虑空间自相关性来进行插值，适用于空间数据的估算。
